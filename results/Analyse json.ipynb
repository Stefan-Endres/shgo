```python
import pandas
import json
%matplotlib inline
rawdata = json.load(open('results.json'))
```

```python
data = []
for problem in rawdata:
    for row in rawdata[problem].values():
        entry = dict(row)
        entry['problem'] = problem
        #if entry['problem'] is not 'All':
        data.append(entry)
```

```python
df = pandas.DataFrame(data, columns=['problem', 'name',  # <-- Needed
                                     'ndim', 'nfev', 'nlmin','nulmin', 'success', 'runtime']).set_index(['problem', 'name'])
```

```python
df.sort_index()
```

```python
fevals = df.nfev.unstack()
```
